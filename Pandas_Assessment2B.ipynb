```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df_1=pd.read_csv("/content/drive/MyDrive/college_1.csv")
df_2=pd.read_csv("/content/drive/MyDrive/college_2.csv")
```

```python
#1.merge those two csv files (after getting as dataframes, get them as a single dataframe)
student_data=pd.concat([df_1,df_2],ignore_index=True)
```

```python
student_data.fillna(0)
```

```python
#2.if Codekata score > 15000(Exceeded expectations.csv)
Exceeded=student_data[(student_data['CodeKata Score'] > 15000)]
new_df=pd.DataFrame(Exceeded)
new_df.to_csv("Exceeded expectations.csv")
```

```python
#3.if 10000<codekata score<15000 (Reached_expectations.csv)
Reached_expectation=(student_data[(student_data['CodeKata Score'] >10000) & (student_data['CodeKata Score'] < 15000)])
new_df=pd.DataFrame(Reached_expectation)
new_df.to_csv("Reached_expectations.csv")
```

```python
#4.if 7000<codekata score<10000 (Needs_Improvement.csv)
Need_improvement=(student_data[(student_data['CodeKata Score'] >7000) & (student_data['CodeKata Score'] < 10000)])
new_df=pd.DataFrame(Need_improvement)
new_df.to_csv("Needs_Improvement.csv")
```

```python
#5.if codekate score < 7000 (Unsatisfactory.csv)
Unsatisfactory=student_data[(student_data['CodeKata Score'] < 7000)]
new_df=pd.DataFrame(Unsatisfactory)
new_df.to_csv("Unsatisfactory.csv")
```

```python
#6.Average of previous week geekions vs this week geekions (i.e Previous Geekions vs CodeKata Score)
Average=student_data[["Previous Geekions","CodeKata Score"]]
avg=px.bar(Average.mean(),title="Average of previous week geekions vs this week geekions")
avg
```

```python
#7. No.of.Students participated
print("Total no.of.Students participated:",student_data["Name"].count())
```

```python
#8.Average completion of python course or my_sql or python english or computational thinking
Avg_Completion=student_data.loc[(student_data['python']==100.0) | (student_data['mysql']==100.0) | (student_data['python_en']==100.0) | (student_data['computational_thinking']==100.0)]
print(round(((len(Avg_Completion.index)/len(student_data.index))*100),2),'%')
```

```python
#9.rising star of the week (top 3 candidate who performed well in that particular week)
top_candidates=student_data.nlargest(3,["Rising"])
top_candidates
#top_candidates["Name"]
```

```python
#10. Shining stars of the week (top 3 candidates who has highest geekions)
top_student=student_data.nlargest(3,["Previous Geekions"])
top_student
#top_student["Name"]
```

```python
#11.Department wise codekata performence (pie chart)
fig=px.pie(student_data['CodeKata Score'],values=student_data['CodeKata Score'],names=student_data["Department"],title="Department wise Codekata Performance")
fig.show()
```

```python
#12.Department wise toppers (horizantal bar graph or any visual representations of your choice)
Dept_topper=px.scatter(student_data,x='Department',y='CodeKata Score',size='CodeKata Score',color='CodeKata Score',hover_data=['Name'])
Dept_topper
```
